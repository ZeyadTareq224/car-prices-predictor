```python
#import pandas
import pandas as pd
```

```python
#getting the data automobile data set
filename = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv'

df = pd.read_csv(filename)
df.head()
```

```python
#as obvious we need to rename some columns to understand the data well
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

#read the data with the new headers
df = pd.read_csv(filename, names = headers)
df.head()
```

```python
#handling missing data 1st:convert '?' to 'nan' for reasons of computational speed and convenience
#we need to import numpy to get the nan
import numpy as np
df.replace('?', np.nan, inplace = True)
df.head()
```

```python
#checking for missing data
missing_data= df.isnull()
missing_data.head()
```

```python
#we can quickly figure out the number of missing values in each column
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
'''
"normalized-losses": 41 missing data
"num-of-doors": 2 missing data
"bore": 4 missing data
"stroke" : 4 missing data
"horsepower": 2 missing data
"peak-rpm": 2 missing data
"price": 4 missing dat

'''
#"normalized-losses": 41 missing data, replace them with mean
#1st calculate the mean of the column
avg_norm_loss = df['normalized-losses'].astype('float').mean(axis = 0)
#2nd Replace "NaN" by mean value in "normalized-losses" column
df['normalized-losses'].replace(np.nan, avg_norm_loss, inplace = True)
```

```python
#"stroke": 4 missing data, replace them with mean
avg_stroke = df['stroke'].astype('float').mean(axis = 0)
df['stroke'].replace(np.nan, avg_stroke, inplace = True)
```

```python
#"bore": 4 missing data, replace them with mean
avg_bore = df['bore'].astype('float').mean(axis = 0)
df['bore'].replace(np.nan, avg_bore, inplace = True)

#"horsepower": 2 missing data, replace them with mean
avg_horsepower = df['horsepower'].astype('float').mean(axis = 0)
df['horsepower'].replace(np.nan, avg_horsepower, inplace = True)

#"peak-rpm": 2 missing data, replace them with mean
avg_peak_rpm = df['peak-rpm'].astype('float').mean(axis = 0)
df['peak-rpm'].replace(np.nan, avg_peak_rpm, inplace = True)
```

```python
#"num-of-doors": 2 missing data, replace them with "four"
df['num-of-doors'].replace(np.nan, 'four', inplace = True)
df['num-of-doors'].value_counts()
```

```python
#"price": 4 missing data, simply delete the whole row
df.dropna(axis = 0, subset = ['price'], inplace = True)

# reset index, because we droped two rows
df.reset_index(drop = True, inplace = True)

#Good! Now, we obtain the dataset with no missing values.
```

```python
#correct the data format
df.dtypes
```

```python
#change bore and stroke to float as they numbers that describe the engine
df[['stroke', 'bore']] = df[['stroke', 'bore']].astype('float')

#change normalized-losses to int as it's all numbers
df['normalized-losses'] = df['normalized-losses'].astype('int')

#change the price and the peak-rpm to floats
df[['price', 'peak-rpm']] = df[['price', 'peak-rpm']].astype('float')
```

```python
df.dtypes
```

```python
#Data Standardization
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df['city-mpg']
df['highway-L/100km'] = 235/df['highway-mpg']
df.head()
```

```python
#Data Normalization
'''
To demonstrate normalization, let's say we want to scale the columns "length", "width" and "height"

Target:would like to Normalize those variables so their value ranges from 0 to 1.

Approach: replace original value by (original value)/(maximum value)

'''
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

```python
#Binning
'''
Binning is a process of transforming continuous numerical variables
into discrete categorical 'bins', for grouped analysis.

'''
#binning the horsepower into low or midum or high

#1st Convert data to correct format
df["horsepower"]=df["horsepower"].astype(int, copy=True)
#We would like 3 bins of equal size bandwidth
#so we use numpy's linspace(start_value, end_value, numbers_generated)
bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
bins
```

```python
#set up group names
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned']= pd.cut(df['horsepower'], bins, labels = group_names, include_lowest = True)

df[['horsepower', 'horsepower-binned']].head()
```

```python
df['horsepower-binned'].value_counts()
```

```python
#dummy variables
#So we can use categorical variables for regression analysis
#get indicator variables and assign it to data frame "dummy_variable_1
dummy_variable_1 = pd.get_dummies(df['fuel-type'])
dummy_variable_1.head()
```

```python
#change cloumns names for clarity
dummy_variable_1.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1"
df = pd.concat([df, dummy_variable_1], axis = 1)

#drop original fuel-type column
df.drop('fuel-type', axis = 1, inplace = True)
```

```python
df.head()
```

```python
#create indicator variable to the column of "aspiration".
dummy_variable_2 = pd.get_dummies(df['aspiration'])
dummy_variable_2.head()
```

```python
dummy_variable_2.rename(columns = {'std':'aspiration-std', 'turbo': 'aspiration-turbo'},inplace=True)
dummy_variable_2.head()
```

```python
#Merge the new dataframe to the original dataframe then drop the column 'aspiration'
df = pd.concat([df, dummy_variable_2], axis = 1)
df.drop('aspiration', axis = 1, inplace = True)
```
